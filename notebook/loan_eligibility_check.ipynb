```python
import mysql.connector
import pandas as pd

# Connect and display existing records
def show_all_applications():
    conn = mysql.connector.connect(
        host="localhost",
        user="root",
        password="tejnk14",
        database="loan_system"
    )
    cursor = conn.cursor()

    query = """
    SELECT 
        c.CustomerID,
        c.Name,
        c.Age,
        ch.CreditScore,
        c.Income,
        l.LoanAmount,
        er.EligibilityStatus
    FROM Customers c
    JOIN CreditHistory ch ON c.CustomerID = ch.CustomerID
    JOIN LoanApplications l ON c.CustomerID = l.CustomerID
    JOIN EligibilityResults er ON l.LoanID = er.LoanID;
    """

    cursor.execute(query)
    rows = cursor.fetchall()
    conn.close()

    columns = ['CustomerID', 'Name', 'Age', 'CreditScore', 'Income', 'LoanAmount', 'EligibilityStatus']
    df = pd.DataFrame(rows, columns=columns)
    print("\nðŸ“‹ Existing Loan Applications with Eligibility Status:")
    print(df)

# To display current records
show_all_applications()
```

```python
#Insert new applicant and display updated records
def insert_and_display(customer_id, name, age, income, history_id, credit_score, loan_id, loan_amount):
    conn = mysql.connector.connect(
        host="localhost",
        user="root",
        password="tejnk14",
        database="loan_system"
    )
    cursor = conn.cursor()

    # Insert new data
    cursor.execute("""
        INSERT INTO Customers (CustomerID, Name, Age, Income)
        VALUES (%s, %s, %s, %s)
    """, (customer_id, name, age, income))

    cursor.execute("""
        INSERT INTO CreditHistory (HistoryID, CustomerID, CreditScore)
        VALUES (%s, %s, %s)
    """, (history_id, customer_id, credit_score))

    cursor.execute("""
        INSERT INTO LoanApplications (LoanID, CustomerID, LoanAmount)
        VALUES (%s, %s, %s)
    """, (loan_id, customer_id, loan_amount))

    conn.commit()

    # Fetch updated records
    query = """
    SELECT 
        c.CustomerID,
        c.Name,
        c.Age,
        ch.CreditScore,
        c.Income,
        l.LoanAmount,
        er.EligibilityStatus
    FROM Customers c
    JOIN CreditHistory ch ON c.CustomerID = ch.CustomerID
    JOIN LoanApplications l ON c.CustomerID = l.CustomerID
    JOIN EligibilityResults er ON l.LoanID = er.LoanID;
    """

    cursor.execute(query)
    rows = cursor.fetchall()
    conn.close()

    columns = ['CustomerID', 'Name', 'Age', 'CreditScore', 'Income', 'LoanAmount', 'EligibilityStatus']
    df = pd.DataFrame(rows, columns=columns)
    print("\nâœ… New Applicant Added. Updated Loan Applications:")
    print(df)

#New applicant details
insert_and_display(
    customer_id=307,
    name='Ishita Kulkarni',
    age=32,
    income=820000,
    history_id=407,
    credit_score=762,
    loan_id=507,
    loan_amount=300000
)

show_all_applications()
```
